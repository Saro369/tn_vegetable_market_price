```python
# Arrange the vegetable price details in a tabular format
import pandas as pd 

df = pd.read_excel('/Users/iamzero/Python/tn_vegetable_market_price/collection_data/raw/tamilNadu_vegetable_market_history_2017-01-01.xlsx')

# Split the Retail Price and Shopping Mall columns into min and max price
df[['min_retail_price', 'max_retail_price']] = df['Retail Price'].str.split('-',expand=True)
df[['min_shopping_mall_parice', 'max_shopping_mall_price']] = df['Shopping Mall'].str.split('-',expand=True)
df['wholesale_price'] = df['Wholesale Price']

# Remove the unwanted columns
df = df.drop(['Retail Price', 'Shopping Mall', 'Wholesale Price'], axis=1)

# Create a new column with the "₹" symbol
df['max_retail_price'] = '₹' + df['max_retail_price']
df['max_shopping_mall_price'] = '₹' + df['max_shopping_mall_price']

# arrange the columns
column_arrangement = ['Vegetable', 'wholesale_price', 'min_retail_price', 'max_retail_price', 'min_shopping_mall_parice', 'max_shopping_mall_price', 'Date']

df = df.reindex(columns=column_arrangement)

# convert the excel file to csv
df.to_csv('/Users/iamzero/Python/tn_vegetable_market_price/collection_data/processed/tn_veg_market_price_history.csv',index=False)



df.head(5)
```
